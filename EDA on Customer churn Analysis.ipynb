1-RelationShip between Monthly charges and Total Charges

```python
sns.lmplot(data=new_df_dummies1,x='MonthlyCharges',y='TotalCharges',fit_reg=False)

plt.title("RelationShip between Monthly charges and Total Charges")
plt.show()
```

Total charges increases as Monthly Charges increases-as expected

2-Churn by Monthly Charges and Total Charges

```python
# Create the first KDE plot
Mth = sns.kdeplot(new_df_dummies1.MonthlyCharges[(new_df_dummies1['Churn']==0)], color='r', fill=True)

# Create the second KDE plot on a new axes object
Mth = sns.kdeplot(new_df_dummies1.MonthlyCharges[(new_df_dummies1['Churn']==1)],ax=Mth, color='blue', fill=True)

# Customize the legend, ylabel, xlabel, and title
plt.legend(['No Churn', 'Churn'], loc='upper right')
plt.ylabel("Density")
plt.xlabel("Monthly Charges")
plt.title('Monthly charges by churn')

plt.show()
```

Insight - Churn is high when Monlthy charges are high

```python
Tot = sns.kdeplot(new_df_dummies1.TotalCharges[(new_df_dummies1['Churn']==0)], color='r', fill=True)

# Create the second KDE plot on a new axes object
Tot = sns.kdeplot(new_df_dummies1.TotalCharges[(new_df_dummies1['Churn']==1)],ax=Tot, color='blue', fill=True)

# Customize the legend, ylabel, xlabel, and title
plt.legend(['No Churn', 'Churn'], loc='upper right')
plt.ylabel("Density")
plt.xlabel("Total Charges")
plt.title('Total charges by churn')

plt.show()
```

Insight - As higher Churner at lower Total Charges

However if we combine the insight of 3 Parameter i.e. Tenure , Monthly Charges & Total Charges then the picture is bit clear :- Highly Monthly Charges at lower tenure result into lower Total charges. Hence all there 3 factor viz Higher Monlthy Charge,Lower tenure and Lower Total Charge are linked to High churn

```python
def uniplot(df,col,title,hue=None):
    sns.set_style('whitegrid')
    sns.set_context('talk')
    plt.rcParams['axes.labelsize']=20
    plt.rcParams['axes.titlesize']=22
    plt.rcParams['axes.titlepad']=30
    
    
    temp=pd.Series(data=hue)
    fig,ax = plt.subplots()
    width = len(df[col].unique()) + 7 + 4*len(temp.unique())
    fig.set_size_inches(width,8)
    plt.xticks(rotation = 45)
    plt.yscale('log')
    plt.title(title)
    ax = sns.countplot(data = df,x= col, order =df[col].value_counts().index,hue=hue,palette='bright')
    
    
    plt.show()
    
```

```python
uniplot(new_df1_taget1,col="Partner",title="Distribution of Gender for Churned",hue='gender')
```

Insight-

-People who don't have a partner are the churner in which Females are ahead
